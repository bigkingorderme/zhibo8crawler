```python
#获取武磊的历史个人数据汇总信息
eval(data.loc[3262]['season_performace'])['total']['总计']
```

```python
#把所有球员的数据汇总信息都提取到data表中
#第一步，历史总出场数据
def get_total_play(sp):
    if sp == '[]':
        return 0
        print(0)
    else:
        return eval(sp)['total']['总计']['出场次数']
        print(eval(sp)['total']['总计']['出场次数'])
        
data['total_play'] = data['season_performace'].apply(get_total_play)
```

```python
#第2步，历史总进球数据
def get_total_goal(sp):
    if sp == '[]':
        return 0
    else:
        return eval(sp)['total']['总计']['进球']
        
data['total_goal'] = data['season_performace'].apply(get_total_goal)
```

```python
#第3步，历史总助攻数据
def get_total_assist(sp):
    if sp == '[]':
        return 0
    else:
        return eval(sp)['total']['总计']['助攻']
        
data['total_assist'] = data['season_performace'].apply(get_total_assist)
```

```python
#先进行进球数据的分析

#data.total_goal.value_counts().sort_index(ascending=False)

col=['age', 'games_played', 
       'habit_foot', 'height',
       'name', 'name_en',
       'nationality',
        'position','team_name', 'update_time', 'value',
       'total_play', 'total_goal',
       'total_assist']
data[col]
```

```python
#goal_analyse = data[col].sort_values(by='total_goal',ascending=False)[:20]
```

```python
#agelist = data_china.groupby(by='age').count()['id']
goal_analyse = data[col].sort_values(by='total_goal',ascending=False)[:20]
bar = (
    Bar()
    .add_xaxis(goal_analyse.name.tolist())
    .add_yaxis("进球数", goal_analyse.total_goal.tolist())
    .reversal_axis()
    .set_global_opts(title_opts=opts.TitleOpts(title="进球数前20的球员"))
)
bar.render_notebook()
```

```python
#中国球员排名前10的
data_china = data[data['nationality']=='中国']
goal_analyse_china = data_china[col].sort_values(by='total_goal',ascending=False)[:10]
goal_analyse_china
```

```python
#合并中国前10和世界前10
goal = goal_analyse[:10].append(goal_analyse_china)

bar = (
    Bar()
    .add_xaxis(goal.name.tolist())
    .add_yaxis("进球数", goal.total_goal.tolist())
    .reversal_axis()
    .set_global_opts(title_opts=opts.TitleOpts(title="中国球员与世界球员的进球数对比"))
)
bar.render_notebook()
```
